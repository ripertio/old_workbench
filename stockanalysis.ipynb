```python

fundamentals = ['returnOnEquity', 'ebitdaMargins', 'trailingPE', '52WeekChange']
df = pd.DataFrame(infos)
df = df.set_index('symbol')
df = df[df.columns[df.columns.isin(fundamentals)]]

```

```python
work_df = df
work_df.loc[work_df['returnOnEquity']<999, 'Score']= 0
work_df['Score'] = work_df['Score'] + work_df['returnOnEquity'].apply(lambda x : 1 if x > 0.2 else 0)
work_df['Score'] = work_df['Score'] + work_df['returnOnEquity'].apply(lambda x : -1 if x < 0.1 else 0)
work_df['Score'] = work_df['Score'] + work_df['ebitdaMargins'].apply(lambda x : 1 if x > 0.12 else 0)
work_df['Score'] = work_df['Score'] + work_df['ebitdaMargins'].apply(lambda x : -1 if x < 0.06 else 0)
work_df['Score'] = work_df['Score'] + work_df['trailingPE'].apply(lambda x : 1 if x < 12 else 0)
work_df['Score'] = work_df['Score'] + work_df['trailingPE'].apply(lambda x : -1 if x > 16 else 0)
work_df['Score'] = work_df['Score'] + work_df['52WeekChange'].apply(lambda x : 1 if x > 0.05 else 0)
work_df['Score'] = work_df['Score'] + work_df['52WeekChange'].apply(lambda x : -1 if x < -0.05 else 0)
# work_df.loc[work_df['ebitdaMargins']>0.12, 'Score']= +1
work_df
```
